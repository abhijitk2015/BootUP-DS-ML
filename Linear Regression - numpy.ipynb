```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
area = [1.2,2.4,5.2,1.6,9.2,7.4,6.2,3.5,1.5,4.8,5.6,7.2,3.5,6.2,6.3]
price= [190,210,480,210,870,750,670,420,170,410,490,670,310,590,650]
x = np.array(area)
y = np.array(price)
```

```python
m = 0.6
c = 0.9
LR = 0.002
for i in range(800):
    # get predictions = yhat
    yhat = m*x + c
    # calculate the error
    cost = np.mean(np.square(y-yhat))
    # Apply gradient descent to minimize the error
    djdm = 2*np.dot((yhat-y),x)
    djdc = 2*np.dot((yhat-y),np.ones(len(y)))
    m = m - LR*djdm
    c = c - LR*djdc
    print(cost)
```

```python
np.ones(5)
```

```python
prediction = x*m+c
plt.scatter(area,price)
plt.plot(area,prediction,c='r')
plt.show()
```
