```python
# which is the best regiment after the trainig finished?
df1.groupby("regiment")['postTestScore'].mean()
```

```python
# which company of which regiment is the best for any operation, once training is finished.
df1.groupby(['regiment','company'])['postTestScore'].mean()
```

```python
# create a column which shows the improvement done by the soldiers during the training
df1['improvement'] = df1['postTestScore'] - df1['preTestScore']
df1.head()
```

```python
# which regiment has made most imporvement during the trainig
df1.groupby('regiment')['improvement'].mean()
```

```python
df1['Cname'] = df1['name'].apply(lambda x:x.upper())
df1
```

```python
df1['id'] = df1[['regiment','name']].apply(lambda x:x['regiment']+"-"+x['name'],axis=1)
df1
```

```python
df = pd.merge(left = df1,right=df2,on='name')
df
```

```python
import datetime
today = datetime.datetime.now()
```

```python
df['date_of_birth'] = pd.to_datetime(df['date_of_birth'])
```

```python
df['Age'] = df['date_of_birth'].apply(lambda x:round((today-x).days/365))
df
```

```python
# create a column which shows the relative imporvement done by a soldier with respect to the
# average improvement of the regiment
# relative improvement = regiment_average - soldier_improvement
df['RI'] = df.groupby('regiment')['improvement'].apply(lambda x:x-x.mean())
df
```
